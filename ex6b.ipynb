```python
import numpy as np
import matplotlib.pyplot as plt
import math

#as the task is to project a function of lambda, we need to define it
Lambda_val = np.arange(1, 101)
l=3

#creating the distributions, using numpy.random
pois = np.random.poisson(l, 10000)
norm = np.random.normal(l, math.sqrt(l), 10000)
diff = 10000*(np.mean(pois)-np.mean(norm))
fractional_error = np.abs(diff) / (np.abs(np.mean(norm)*10000))

fr_lambda=[]
for l in range(1,101):
    pois = np.random.poisson(l, 10000)
    norm = np.random.normal(l, math.sqrt(l), 10000)
    diff = 10000*(np.mean(pois)-np.mean(norm))
    fr_error = np.abs(diff) / (np.abs(np.mean(norm)*10000))
    fr_lambda.append(fr_error)

plt.plot(Lambda_val,fr_lambda)

plt.show()
```
